```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
# Import the dataset.
df = pd.read_csv('/Users/guest/Desktop/Thinkful/Pre-Course/Lesson 2/What Do You See?/Un_data_CO2.csv')
```

```python
# Create dataframes representing the curve of CO2 emissions since 1990 by France and the United Kingdom.
france_emissions = df.loc[lambda df: df['country'] == 'France', :]
uk_emissions = df.loc[lambda df: df['country'] == 'United Kingdom', :]
us_emissions = df.loc[lambda df: df['country'] == 'United States of America', :]
```

```python
# Plot the curves for comparison.
plt.scatter(x = france_emissions['year'], y = france_emissions['kt_co2'], color='purple', alpha=0.5)
plt.xlim(1990, 2014)
plt.scatter(x = uk_emissions['year'], y = uk_emissions['kt_co2'], color='red', alpha=0.5)
plt.xlim(1990, 2014)
plt.title('Comparison of CO2 Emissions', fontsize=16)
plt.xlabel('Year (1990-2014)', fontsize=12) # Label the X axis
plt.ylabel('CO2 Emissions (in kt)', fontsize=12) # Label the Y axis
france = mpatches.Patch(color='purple', alpha=0.5, label = 'France') #Label France
uk = mpatches.Patch(color='red', alpha=0.5, label = 'United Kingdom') # Label the UK
plt.legend(handles=[france, uk])
plt.show()
```

```python
# In the chart above, CO₂ emissions from 1990 to 2014 are charted for France and the United Kingdom. I selected
# these two states for comparison because they have similar population sizes. While the curves for the two lines are
# extremely similar (note the spike in 1996, the dips in 2009 and 2011, and the overall downward trend), the UK's
# CO₂ emissions are considerably higher.
```

```python
# Plot histograms for comparison.
plt.hist(x = france_emissions['kt_co2'], normed=True, color='purple', bins = 6, alpha=.5) 
plt.hist(x = uk_emissions['kt_co2'], normed=True, color='red', bins = 6, alpha=.5)
plt.title('Distributions of CO2 Emission Levels', fontsize=16)
plt.xlabel('CO2 Emissions (in kt)', fontsize=12)
france = mpatches.Patch(color='purple', alpha=0.5, label = 'France') #Label France
uk = mpatches.Patch(color='red', alpha=0.5, label = 'United Kingdom') # Label the UK
plt.legend(handles=[france, uk])
plt.show()
```

```python
# In the above histograms, CO₂ emissions for France and the United Kingdom are compared. Both distributions have a
# strong positive skew, indicating higher frequencies of higher annual emissions levels from 1990-2014 for both states.
# As observed in the first chart, the United Kingdom's emissions levels are higher than France's.
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
france_emissions.boxplot(column='kt_co2')
plt.title('Boxplot of French CO2 Emissions (1990-2014)', fontsize=12)
plt.ylabel('CO2 Emissions (in kt)', fontsize=12)

plt.subplot(1, 2, 2)
uk_emissions.boxplot(column='kt_co2',)
plt.title('Boxplot of UK CO2 Emissions (1990-2014)', fontsize=12)
plt.ylabel('CO2 Emissions (in kt)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# The boxplots above show the range of CO2 emissions by year for France and the United Kingdom. There is one outlier
# in the French boxplot, and there are four outliers in the United Kingdom boxplot. 
```
