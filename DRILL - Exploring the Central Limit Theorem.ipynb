```python
import numpy as np
import pandas as pd
import scipy
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Recreate the populations from the previous assignment.
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)
```

```python
# Recreate the samples from the previous assignment.
p1_n100 = np.random.choice(pop1, 100, replace=True)
p2_n100 = np.random.choice(pop2, 100, replace=True)
```

```python
# Calculate the mean and standard deviation of each sample.
# Create histograms for each.
plt.hist(p1_n100, alpha=0.5, label='population 1 (n = 100)')
plt.legend(loc='upper right')
plt.show()

print('mean = ' + str(p1_n100.mean()))
print('standard deviation = ' + str(p1_n100.std()))

plt.hist(p2_n100, alpha=0.5, label='population 2 (n = 100)', color='red')
plt.legend(loc='upper right')
plt.show()

print('mean = ' + str(p2_n100.mean()))
print('standard deviation = ' + str(p2_n100.std()))
```

```python
# Increase the size of the samples from 100 to 1000.
p1_n1000 = np.random.choice(pop1, 1000, replace=True)
p2_n1000 = np.random.choice(pop2, 1000, replace=True)
```

```python
# Calculate the mean and standard deviation of each sample.
# Create histograms for each.


plt.hist(p1_n1000, alpha=0.5, label='population 1 (n = 1000)')
plt.legend(loc='upper right')
plt.show()
print('mean = ' + str(p1_n1000.mean()))
print('standard deviation = ' + str(p1_n1000.std()))

plt.hist(p2_n1000, alpha=0.5, label='population 2 (n = 1000)', color='red')
plt.legend(loc='upper right')
plt.show()
print('mean = ' + str(p2_n1000.mean()))
print('standard deviation = ' + str(p2_n1000.std()))
```

```python
# Decrease the size of the samples from 1000 to 20.
p1_n20 = np.random.choice(pop1, 20, replace=True)
p2_n20 = np.random.choice(pop2, 20, replace=True)
```

```python
# Calculate the mean and standard deviation of each sample.
# Create histograms for each.

plt.hist(p1_n20, alpha=0.5, label='population 1 (n = 20)')
plt.legend(loc='upper right')
plt.show()
print('mean = ' + str(p1_n20.mean()))
print('standard deviation = ' + str(p1_n20.std()))

plt.hist(p2_n20, alpha=0.5, label='population 2 (n = 20)', color='red')
plt.legend(loc='upper right')
plt.show()
print('mean = ' + str(p2_n20.mean()))
print('standard deviation = ' + str(p2_n20.std()))
```

```python
# Calculate the mean and standard deviation of each sample.
print("Population 1:")
print('\nmean (n=20) = ' + str(p1_n20.mean()))
print('mean (n=100) = ' + str(p1_n100.mean()))
print('mean (n=1000) = ' + str(p1_n1000.mean()))

print('\nstandard deviation (n=20) = ' + str(p1_n20.std()))
print('standard deviation (n=100) = ' + str(p1_n100.std()))
print('standard deviation (n=1000) = ' + str(p1_n1000.std()))

print("\n\n\nPopulation 2:")
print('\nmean (n=20) = ' + str(p2_n20.mean()))
print('mean (n=100) = ' + str(p2_n100.mean()))
print('mean (n=1000) = ' + str(p2_n1000.mean()))

print('\nstandard deviation (n=20) = ' + str(p2_n20.std()))
print('standard deviation (n=100) = ' + str(p2_n100.std()))
print('standard deviation (n=1000) = ' + str(p2_n1000.std()))

# What values change, and what remain the same?
print('\n\n\nThe means stayed similar across all samples, but the standard deviations decreased in inverse proportion to sample size. This makes sense – the larger samples should have less variance than their smaller counterparts.')
```
